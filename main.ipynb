```python
# tratamento de dados

# colunas
df_vendas[ "Loja" ]

# excluir uma coluna
df_analise = df_vendas.drop(columns=["Data_Base"])
```

```python
display(df_analise)
```

```python
# valores nulos da base de dados - linha não executada
df_analise = df_analise.dropna() # exclui a linha que tem valor vazio
```

```python
# trocar os valores vazios (substitui o NaN da loja por venda online)
df_analise["Loja"] = df_analise["Loja"].fillna("Online")
```

```python
# tipos de dados errados (data)
display(df_analise["Data"])
```

```python
df_analise["Data"] = pd.to_datetime(df_analise["Data"], format="%Y-%m-%d")
```

```python
# padronização de dados
df_analise["Loja"] = df_analise["Loja"].str.strip() # remove espaços inuteis
df_analise["Loja"] = df_analise["Loja"].str.title() # padroniza com a primeira letra maiscula
```
